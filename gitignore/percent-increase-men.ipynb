```python
import pandas as pd
import json

# Load JSON file
with open('manav_matches.json', 'r') as file:
    data = json.load(file)

df = pd.DataFrame(data)

df
```

```python
df['match_time'] = df['match'].apply(lambda x: pd.to_datetime(x[0]['timestamp']) if isinstance(x, list) and len(x) > 0 else None)

#Filtering to all non-na values
df = df[df['match_time'].notna()]

#Extracting date from full timestamp including hour
df['match_date'] = df['match_time'].dt.date

#Extracting month from date
df['match_month'] = df['match_time'].dt.to_period('M')

#Grouping by month
monthly_counts = df.groupby('match_month').size().reset_index(name='count')


print(monthly_counts)
monthly_counts.to_csv("manav-monthly-matches.csv")
```
